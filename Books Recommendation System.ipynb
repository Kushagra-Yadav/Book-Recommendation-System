# Popularity Based Filtering

```python
# Take the top 50 books with maximum average rating considering only those books rated by more than equal to 250 
```

```python
books_with_ratings=books.merge(ratings,on='ISBN')
books_with_ratings
```

```python
numRating=books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
numRating.rename(columns={'Book-Rating':'Num-Rating'},inplace=True)
numRating
```

```python
avgRating=books_with_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avgRating.rename(columns={'Book-Rating':'Avg-Rating'},inplace=True)
avgRating
```

```python
popularity_res=numRating.merge(avgRating,on='Book-Title')
popularity_res=popularity_res[popularity_res['Num-Rating']>=250]
popularity_res
```

```python
popularity_res=popularity_res.sort_values('Avg-Rating',ascending=False).head(50)
print(popularity_res.shape)
popularity_res
```

```python
popularity_res=popularity_res.merge(books,on='Book-Title')
popularity_res.shape
```

```python
popularity_res=popularity_res.drop_duplicates('Book-Title')
print(popularity_res.shape)
popularity_res
```

```python
popularity_res=popularity_res[['Book-Title','Book-Author','Num-Rating','Avg-Rating','Image-URL-M','Publisher']]
popularity_res
```
